```python
import matplotlib.pyplot as plt
import pandas as pd

# Creamos el DataFrame de ejemplo
data = {
    "Concepto": ["Renta", "Super", "Gasolina", "Netflix", "Luz", "Agua", "Internet", "Cena"],
    "Monto": [2750, 900, 1200, 199, 600, 400, 500, 1500],
    "Fecha": [
        "2025-08-01", "2025-08-02", "2025-08-03", "2025-08-04",
        "2025-08-05", "2025-08-06", "2025-08-07", "2025-08-08"
    ]
}
df = pd.DataFrame(data)

# 1) Filtrar gastos mayores a 1000
f1 = df[df['Monto'] > 1000]
print("ðŸ‘‰ Gastos mayores a 1000:\n", f1)

# 2) Graficar en barras esos gastos
plt.bar(f1['Concepto'], f1['Monto'])
plt.title("Gastos mayores a 1000")
plt.xlabel("Concepto")
plt.ylabel("Monto")
plt.show()

# 3) Otra grÃ¡fica (ejemplo: pastel de todos los gastos)
plt.pie(df['Monto'], labels=df['Concepto'], autopct='%1.1f%%')
plt.title("DistribuciÃ³n de todos los gastos")
plt.show()
```

```python
import pandas as pd

# Datos de ejemplo
data = {
    "Concepto": ["Renta", "Super", "Gasolina", "Netflix", "Luz", "Agua", "Internet", "Cena"],
    "Monto": [2750, 900, 1200, 199, 600, 400, 500, 1500],
    "Fecha": [
        "2025-08-01", "2025-08-02", "2025-08-03", "2025-08-04",
        "2025-08-05", "2025-08-06", "2025-08-07", "2025-08-08"
    ]
}

df = pd.DataFrame(data)

# ðŸ‘‰ Usando .loc para filtrar con condiciones
# Ejemplo: gastos mayores a 1000
gastos_mayores = df.loc[df["Monto"] > 1000]
print("Gastos mayores a 1000:\n", gastos_mayores)

# Otro ejemplo: seleccionar solo Concepto y Monto de las filas donde Concepto sea 'Renta' o 'Cena'
seleccion = df.loc[df["Concepto"].isin(["Renta", "Cena"]), ["Concepto", "Monto"]]
print("\nFiltrado por Concepto especÃ­fico:\n", seleccion)

# Ejemplo: rango de fechas usando loc
rango_fechas = df.loc[(df["Fecha"] >= "2025-08-02") & (df["Fecha"] <= "2025-08-06")]
print("\nGastos entre 02 y 06 de agosto:\n", rango_fechas)
```
