```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Concepto": ["Renta", "Super", "Gasolina", "Super", "Gasolina", "Luz", "Agua", "Renta","Agua"],
    "Monto": [2750, 900, 1200, 1100, 800, 600, 400, 2750,150],
    "Mes": ["Agosto", "Agosto", "Agosto", "Septiembre", "Septiembre", "Agosto", "Septiembre", "Septiembre","Septiembre"]
}

df = pd.DataFrame(data)

# Histograma de montos
df["Monto"].plot(kind="hist", bins=5, title="Distribución de montos")
plt.xlabel("Monto")
plt.show()
```

```python
df.boxplot(column="Monto", by="Mes")
plt.title("Distribución de montos por mes")
plt.suptitle("")  # para limpiar el título automático
plt.show()
```
