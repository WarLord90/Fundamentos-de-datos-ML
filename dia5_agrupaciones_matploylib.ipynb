```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Concepto": ["Renta", "Super", "Gasolina", "Super", "Gasolina", "Luz", "Agua", "Renta","Agua"],
    "Monto": [2750, 900, 1200, 1100, 800, 600, 400, 2750,150],
    "Mes": ["Agosto", "Agosto", "Agosto", "Septiembre", "Septiembre", "Agosto", "Septiembre", "Septiembre","Septiembre"]
}

df = pd.DataFrame(data)
```

```python
total_por_concepto = df.groupby("Concepto")["Monto"].sum()

total_por_concepto.plot(kind="bar", title="Gasto total por concepto")
plt.xlabel("Concepto")
plt.ylabel("Monto")
plt.show()
```

```python
total_por_mes = df.groupby("Mes")["Monto"].sum()

total_por_mes.plot(kind="pie", autopct="%1.1f%%", title="Distribuci√≥n de gasto por mes")
plt.ylabel("")  # quitar etiqueta del eje Y
plt.show()
```

```python
detalle = df.groupby(["Mes", "Concepto"])["Monto"].sum().unstack()

detalle.plot(kind="bar", stacked=True, title="Gastos por concepto en cada mes")
plt.xlabel("Mes")
plt.ylabel("Monto")
plt.show()
```

```python
promedio_concepto = df.groupby("Concepto")["Monto"].mean()

promedio_concepto.plot(kind="barh", color="orange", title="Promedio de gasto por concepto")
plt.xlabel("Promedio")
plt.ylabel("Concepto")
plt.show()
```
