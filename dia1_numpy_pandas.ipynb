```python
import pandas as pd

df = pd.DataFrame({
    'Empleado': ['Ana','Luis','Ana','Marta','Luis','Ana','Marta'],
    'Ventas':   [3200, 5100, 4300, 6100, 2900, 5200, 4800],
    'Mes':      ['Enero','Enero','Febrero','Febrero','Marzo','Marzo','Marzo']
})

df
```

```python
df[df['Ventas'] > 5000]
```

```python
ventas_mes = df.groupby('Mes', as_index=False)['Ventas'].sum()
ventas_mes
```

```python
import matplotlib.pyplot as plt

plt.bar(ventas_mes['Mes'], ventas_mes['Ventas'])
plt.title('Ventas por Mes')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.show()
```

```python
import pandas as pd

# Ingresos
sueldo_quincenal = 13087

# Gastos por quincena
quincena1 = {
    'prestamo': 7000,
    'comida': 1200,
    'at&t': 916,
    'gasolina': 1000,
    'casetas': 600,
    'google': 50,
    'chatgpt': 400,
    'banamex': 500,
    'alan': 250,
    'almohada': 329
}

quincena2 = {
    'prestamo': 7000,
    'comida': 1000,
    'cable': 717,
    'gasolina': 500,
    'renta': 2800,
    'casetas': 300,
    'youtube': 160,
    'meli': 130
}

# Convertir a DataFrames
df1 = pd.DataFrame(list(quincena1.items()), columns=['Concepto','Monto'])
df1['Quincena'] = 'Q1'

df2 = pd.DataFrame(list(quincena2.items()), columns=['Concepto','Monto'])
df2['Quincena'] = 'Q2'

df = pd.concat([df1, df2])

# Totales
totales = df.groupby('Quincena')['Monto'].sum().reset_index()
totales['Sueldo'] = sueldo_quincenal
totales['Resta'] = totales['Sueldo'] - totales['Monto']

display(df)
display(totales)
```

```python
import matplotlib.pyplot as plt
# Gr치fico comparando gasto vs resta por quincena
plt.bar(totales['Quincena'], totales['Monto'], label='Gasto')
plt.bar(totales['Quincena'], totales['Resta'], label='Resta', alpha=0.7)

plt.title('Gasto y Resta por Quincena')
plt.xlabel('Quincena')
plt.ylabel('Monto')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# posiciones en el eje X
x = np.arange(len(totales['Quincena']))

# ancho de cada barra
width = 0.35  

fig, ax = plt.subplots()

# barras de gasto
ax.bar(x - width/2, totales['Monto'], width, label='Gasto', color='tomato')

# barras de resta
ax.bar(x + width/2, totales['Resta'], width, label='Resta', color='seagreen')

# etiquetas y t칤tulo
ax.set_title('Gasto vs Resta por Quincena')
ax.set_xlabel('Quincena')
ax.set_ylabel('Monto')
ax.set_xticks(x)
ax.set_xticklabels(totales['Quincena'])
ax.legend()

plt.show()
```
